## Using learnt representations
In practice, very few people train an entire Convolutional Network from scratch (with random initialization), because it is relatively rare to have a dataset of sufficient size. Instead, it is common to pretrain a ConvNet on a very large dataset (e.g. ImageNet, which contains 1.2 million images with 1000 categories), and then use the ConvNet either as an initialization or a fixed feature extractor for the task of interest.


We'll train a model to classify ants and bees. We have about 120 training images each for ants and bees. There are 75 validation images for each class. Usually, this is a very small dataset to generalize upon, if trained from scratch. Since we are using transfer learning, we should be able to generalize reasonably well.

This dataset is a very small subset of imagenet.




```python
# Import packages
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, SubsetRandomSampler
import torch.nn as nn
from torch import optim
import torch.nn.functional as F

import matplotlib.pyplot as plt
import numpy as np
import time
import os
```

```python
# Device configuration (whether to run on GPU or CPU)
device = 'cpu' #torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Download and extract dataset
!wget https://download.pytorch.org/tutorial/hymenoptera_data.zip
!unzip -q hymenoptera_data.zip
```

```python
# Data augmentation and normalization for training
# Just normalization for validation
train_transform = transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
val_transform = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

data_dir = './hymenoptera_data'
train_dataset = ImageFolder(os.path.join(data_dir, 'train'), train_transform)
val_dataset = ImageFolder(os.path.join(data_dir, 'val'), val_transform)
train_dataloader = DataLoader(train_dataset, batch_size=4,shuffle=True, num_workers=2)
val_dataloader = DataLoader(val_dataset, batch_size=4,shuffle=True, num_workers=2)
class_names = train_dataset.classes
```

```python
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated


# Get a batch of training data
inputs, classes = next(iter(train_dataloader))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

```python
def train_model(model, train_dataloader, criterion, optimizer, num_epochs=25):

    for epoch in range(num_epochs):
        print(f'Epoch {epoch}/{num_epochs - 1}')
        print('-' * 10)
        model.train()
        running_loss = 0.0
        running_corrects = 0

        # Iterate over data.
        for i, (inputs, labels) in enumerate(train_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            # zero the parameter gradients
            optimizer.zero_grad()

            # forward
            # track history if only in train
            with torch.set_grad_enabled(True):
              outputs = model(inputs)
              _, preds = torch.max(outputs, 1)
              loss = criterion(outputs, labels)

              # backward + optimize only if in training phase
              loss.backward()
              optimizer.step()

            # statistics
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(train_dataloader.dataset)
        epoch_acc = running_corrects.double() / len(train_dataloader.dataset)

        print(f'Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

        print()
```

```python
def visualize_model(model, num_images=10):
    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure()

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(val_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images//2, 2, images_so_far)
                ax.axis('off')
                ax.set_title(f'predicted: {class_names[preds[j]]}')
                imshow(inputs.cpu().data[j])

                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)
```

```python
from torchvision import models

model_ft = models.resnet18(pretrained=True)
num_ftrs = model_ft.fc.in_features
# Here the size of each output sample is set to 2.
# Alternatively, it can be generalized to nn.Linear(num_ftrs, len(class_names)).
model_ft.fc = nn.Linear(num_ftrs, 2)

model_ft = model_ft.to(device)

loss_func = nn.CrossEntropyLoss()

# Observe that all parameters are being optimized
optimizer = optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)
```

```python
num_epochs = 5
train_model(model_ft, train_dataloader, loss_func, optimizer, num_epochs)
```

```python
visualize_model(model_ft)
```

**1**.    The **significance of using data augmentations** like **resize**, **crop**, **flip**, **rotation**, etc., on training data in machine learning—especially in computer vision—lies in making the model **more robust, accurate, and generalizable**. Here's why:

---

### 🔁 **1. Increases Dataset Diversity**
- Augmentations simulate various real-world scenarios.
- Helps expose the model to different versions of the same image without actually needing more data.

**Example:** A bee in the center vs. a bee in the corner – cropping can help the model recognize both.

---

### 🧠 **2. Prevents Overfitting**
- Without augmentation, the model might memorize the training data (especially in small datasets).
- Augmentations add noise and variability, forcing the model to learn **general patterns** instead of memorizing specific images.

---

### 🔍 **3. Improves Generalization to New Data**
- Models trained with augmentations usually perform better on unseen/test images.
- This is because they’ve learned to recognize objects despite changes in size, position, orientation, lighting, etc.

---

### 🧰 **4. Makes the Model Invariant to Certain Transformations**
- **Resize:** Makes input dimensions consistent for the model.
- **Crop:** Helps with localization and teaches the model to recognize partial objects.
- **Flip/Rotate:** Ensures the model isn't biased to object orientation.

---

### 🧪 Summary:
| Augmentation | Purpose |
|--------------|---------|
| Resize       | Standardizes image size |
| Crop         | Teaches spatial variance |
| Flip/Rotate  | Adds orientation diversity |
| Color Jitter | Helps with lighting/color changes |
| Normalize    | Scales pixel values, helps training stability |

---

So overall, augmentations make the training **more realistic**, **more challenging**, and **more effective**—especially when data is limited.



**2**   If you **don't use a pretrained ResNet model** (i.e., set `pretrained=False`) and train it **from scratch** on a small dataset like **Ants vs Bees**, here’s what typically happens:

---

### 🚫 **Performance Without Pretrained Weights**

#### 🔻 **1. Lower Accuracy**
- Accuracy often drops significantly—usually in the range of **50–65%** depending on the number of epochs, optimizer settings, and augmentation used.
- The model struggles to distinguish between ants and bees effectively in the early stages.

#### 🐌 **2. Slower Convergence**
- It takes more epochs for the model to start learning meaningful patterns.
- Since the model starts from random weights, learning low-level features (edges, corners, etc.) from scratch is inefficient.

#### 🎯 **3. High Risk of Overfitting**
- With only a few hundred training images and no prior knowledge, the model often **memorizes** the training set.
- Generalization to validation/test data suffers.

---

### 📈 Example (based on typical results from PyTorch or similar frameworks):

| Model Type | Accuracy (after ~10 epochs) | Training Time | Remarks |
|------------|------------------------------|----------------|---------|
| **Pretrained=True** | ~90–95% | Faster | Leverages ImageNet knowledge |
| **Pretrained=False** | ~55–65% | Slower | Struggles with small dataset |

---

### 🧠 Why Is This Happening?

A pretrained ResNet has already learned:
- Low-level features: edges, textures.
- Mid-level features: shapes, patterns.
- High-level features: object-specific abstractions.

By contrast, training from scratch requires:
- Learning all of these with **very limited data**, which is extremely hard and inefficient.



**3**   Great question! Let’s break it down:

---

### 🧠 **If the ResNet Model Was Pretrained on a Dataset Significantly Different Than Ants vs Bees...**

You would **still get reasonably good performance** in most cases — **better than training from scratch**, even if the pretraining dataset is different.

---

### 💡 **Why It Still Works:**

#### ✅ **1. Early Layers Learn Universal Features**
- The initial layers of a CNN like ResNet learn **generic patterns** such as edges, textures, and shapes.
- These are useful across **almost any image dataset**, even if the content is different.

#### 🔄 **2. Transfer Learning Still Helps**
- Even if the pretraining dataset (say, medical images, satellite imagery) is very different, the **learned weights** still provide a strong starting point.
- You only need to **fine-tune the later layers** to adapt to your target dataset.

#### 🔧 **3. Fine-tuning Makes It Work**
- If the pretraining dataset is very different, you might want to:
  - Unfreeze more layers during training.
  - Use a **lower learning rate** for pretrained layers and a higher one for new layers.

---

### 📉 What’s the Trade-Off?

| Pretraining Dataset | Expected Performance | Notes |
|----------------------|----------------------|-------|
| **Very Similar (e.g., insects)** | ⭐⭐⭐⭐⭐ (Best) | Fast convergence, high accuracy |
| **Generic (e.g., ImageNet)** | ⭐⭐⭐⭐ | Still very good |
| **Very Different (e.g., X-ray, satellite)** | ⭐⭐⭐ | Decent with fine-tuning |
| **None (train from scratch)** | ⭐ | Poor without lots of data |

---

### 🎯 Final Verdict:
Even if the pretraining data is **very different**, using a pretrained model still gives you a **better starting point**, improves **generalization**, and typically results in **higher accuracy** than starting from scratch.
